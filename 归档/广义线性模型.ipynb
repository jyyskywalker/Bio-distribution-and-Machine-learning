# 广义线性模型
线性模型是一类广泛应用于机器学习领域的预测模型。它是使用输入数据集的特性的线性函数进行建模，并对结果进行预测的方法。
## 线性模型的基本概念
线性模型不是某一个模型，而是一类模型。常用的广义线性模型包括：线性回归、岭回归、套索回归、逻辑回归和线性SVC等。
### 线性模型的一般公式
在回归分析中，线性模型的一般预测公式如下:
$$ \hat{y} = w_0x_0 + w_1x_1 + \cdots + w_nx_n + b $$
其中：$ w $和$ b $为模型的参数，$ \hat{y} $为模型对于数据结果的预测值。
- 注意：$ \hat{y} $ 读作“y hat”，为模型对于数据结果的预测值

```python

# ^y = w1*x1+w2*x2+w2*x2+wn*xn+b
# ^y (y hat) 代表y的估计值

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-5,5,101)
#print(x)
y=0.5*x+3
#print(y)
plt.plot(x,y,c='#99cc33')
plt.title('Straight Line')
plt.show()
```

### 线性模型的图形表示
两个点确定一条直线。假设两个点为(1,3)和(4,5)，我们可以画一条直线穿过这两个点，并计算出其方程。

```python
# 导入线性回归模型
from sklearn.linear_model import LinearRegression
# 横坐标
X=[[1],[4]]
# 纵坐标
y=[3,5]
# 线性模型拟合
lr=LinearRegression().fit(X,y)

z = np.linspace(0,5,21)
plt.scatter(X,y,c='#cc3399',s=128)
# 画直线
plt.plot(z,lr.predict(z.reshape(-1,1)),c='k')
plt.show()
print('直线方程: y = {:.3f}'.format(lr.coef_[0]),'x','+ {:.3f}'.format(lr.intercept_))
```

**[结果分析]** 得到的结果是穿过两个数据点的直线

```python
# 3个点的情况，加个(3,3)
# 导入线性回归模型
from sklearn.linear_model import LinearRegression
# 横坐标
X=[[1],[4],[3]]
# 纵坐标
y=[3,5,3]
# 线性模型拟合
lr=LinearRegression().fit(X,y)
z = np.linspace(0,5,21)
plt.scatter(X,y,c='#3399cc',s=128)
# 画直线
plt.plot(z,lr.predict(z.reshape(-1,1)),c='k')
plt.show()
print('直线方程: y = {:.3f}'.format(lr.coef_[0]),'x','+ {:.3f}'.format(lr.intercept_))
```

**[结果分析]** 直线没有穿过任何一个点，而是位于和3个点的距离相加最小的位置

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

X, y = make_regression(n_samples=50, n_features=1, n_informative=1,
                       noise=50,random_state=1)
lr = LinearRegression()
lr.fit(X,y)
z = np.linspace(-5,5,101).reshape(-1,1)
plt.scatter(X,y,c='#33cc99',s=60)
plt.plot(z, lr.predict(z),c='k')
plt.title('Linear Regression')
plt.show()

print('直线方程: y = {:.3f}'.format(lr.coef_[0]),'x','+ {:.3f}'.format(lr.intercept_))
print('直线系数: {:.3f}'.format(lr.coef_[0]))
print('直线截距: {:.3f}'.format(lr.intercept_))
```

**[结果分析]** 我们可以看出，这条直线距离这50个点的距离之和是最小的。这便是一般线性模型的原理。
- 注意：coef_和intercept_都以下划线结尾，这是sklearn的一个特点，它总是以下划线作为子训练数据集的属性结尾，以便将其与用户参数区分

### 线性模型的特点
上述都为特征数量为1的数据集，此时预测模型是一条直线，特征数量为2时是一个平面，更多则是高纬度超平面。
与KNN相比，线性模型有其局限性，很多数据没有体现在直线上，因为使用线性模型的前提是假设目标$ y $是数据特征的线性组合。
当然上述都是1维数据集，如果增加特征变量，线性模型就会显得十分强大，如果特征变量大于数据点的数量，线性模型可以做出完美预测。
线性模型之间的区别在于如何确定模型参数$ w $和$ b $，以及如何控制模型复杂度。
